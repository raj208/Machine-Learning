```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("onlinefoods.csv")
```

```python
df.head()
```

```python
sex = pd.get_dummies(df['Gender'], drop_first = True)
```

```python
sex
```

```python
df.drop(['Gender','Occupation','Monthly Income','Educational Qualifications'],axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = pd.concat([df,sex], axis = 1)
```

```python
df.head()
```

```python
Martial = pd.get_dummies(df['Marital Status'], drop_first = True)
```

```python
Martial
```

```python
df.drop('Marital Status', axis = 1, inplace = True)
```

```python
Output = pd.get_dummies(df['Output'], drop_first = True)
Feedback = pd.get_dummies(df['Feedback'], drop_first = True)
Target = pd.get_dummies(df['Target'], drop_first = True)
```

```python
df.drop(['Output','Feedback','Target'], axis = 1, inplace = True)
```

```python
df = pd.concat([df,Output,Feedback,Target], axis = 1)
```

```python
df.info()
```

```python
df.rename(columns={df.columns[8]: 'Output'}, inplace=True)
df.head()
```

```python
df.rename(columns={'Target': 'Output'}, inplace=True)
```

```python

column_labels = df.columns.tolist() 

# Find the index of the first occurrence of the column 'Output'
first_occurrence_index = column_labels.index('Output')

# Rename only the first occurrence of the column 'Output'
new_column_labels = column_labels[:first_occurrence_index] + ['Target'] + column_labels[first_occurrence_index+1:]
df.columns = new_column_labels
```

```python
df.head()
```

```python
df.rename(columns={'Target': 'Output', 'Output':'Target'}, inplace=True)
```

```python
df.drop('Pin code', axis = 1, inplace = True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Target', axis = 1)
```

```python
y = df['Target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, pred)
```

```python
print(classification_report(y_test, pred))
```
